```python
d=dict()
def add_to_dict(words,d):
    i=len(d)
    for word in words:
        if word not in d :
            d[word]=i
            i+=1
```

```python
df['text']=df['text'].apply(ast.literal_eval)
```

```python
list_physical_devices('GPU')
```

```python
df['text']
```

```python
for ind in df.index:
    add_to_dict(df['text'][ind],d)
```

```python

def sentence_vectorization(list_of_words):
    list=[]
    for i in list_of_words:
      list.append(d[i])
    return list
```

```python
df['text']=df['text'].apply(sentence_vectorization)
```

```python
del stop_words
```

```python
max_len=0
for i in df['text'].tolist():
    if len(i)>max_len:
        max_len=len(i)
```

```python
listOfKeys=list(d.keys())
```

```python
padded = pad_sequences(df['text'].tolist(), maxlen=max_len,value=len(d))
count=0
for ind in df.index:
    df.at[ind,'text']=padded[count]
    count+=1
```

```python
d["PAD"]=len(d)
```

```python
embedding_matrix=np.zeros((len(d),300))
for i in range(0,embedding_matrix.shape[0]):
  try:
    embedding_matrix[i]=w2v[listOfKeys[i]]
  except:
    pass
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['sentiment'], test_size=0.2, random_state=42)
```

```python
del w2v
```

```python
del df
```

```python
model = Sequential()
model.add(layers.Embedding(input_dim=embedding_matrix.shape[0],
                           output_dim=embedding_matrix.shape[1],
                           weights=[embedding_matrix],
                           trainable=True))
model.add(Dropout(0.2))
model.add(SimpleRNN(300))
model.add(Dense(1))
model.summary()
```

```python
model.compile(
    loss=BinaryCrossentropy(from_logits=True),
    optimizer="Adam",
    metrics=["accuracy"],
)
```

```python
del count, d , embedding_matrix,ind,listOfKeys,padded,stopwords
```

```python

X_train = np.array(X_train.values.tolist())
X_test = np.array(X_test.values.tolist())
```

```python
callback = EarlyStopping(monitor='loss', patience=3)
```

```python
model.fit(
    X_train, y_train, batch_size=512, epochs=5,
    callbacks=[callback]
)
```

```python
y_pred=model.predict(X_test)
```

```python
del max_len,callback,i
```

```python
del X_train,y_train
```

```python
y_pred
y_pred[y_pred >= 0.5] = 1
y_pred[y_pred < 0.5] = 0
accuracy_score(y_test,y_pred)
report=classification_report(y_test, y_pred,target_names=['negetive','posetive'])
```
